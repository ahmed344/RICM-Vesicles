```python
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
normal, small, blocked, bad = [], [], [], []
for name in [name for name in os.listdir("./20201218_Integrin/results")]:
    if "Bad" in name:
        bad.append(float(name[-5:]))
    elif "Blocked" in name:
        blocked.append(float(name[-5:]))
    elif "Small" in name:
        small.append(float(name[-5:]))
    else:
        normal.append(float(name[-5:]))
```

```python
plt.figure(figsize=(5,15))
plt.scatter(['bad']*len(bad), bad)
plt.scatter(['small']*len(small), small)
plt.scatter(['normal']*len(normal), normal)
plt.scatter(['blocked']*len(blocked), blocked)
plt.xlabel("Cells")
plt.ylabel("Height")
plt.yticks(np.linspace(63,73.5,22))
plt.savefig('data_scatter')
plt.show()
```
